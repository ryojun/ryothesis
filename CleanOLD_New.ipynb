```python
def cutUrl(input):
    postion = input.split('/')
    for x in postion:
        if len(x) > 50:
            posis = x
            findName = posis.split('.')
            return findName[0]

def obbEvev(input):
    if int(input)%2 == 0 :
        return 'even'
    elif int(input)%2 != 0:
        return 'odd'
    
def likePosition(likeInput,Like):
    #q meaning quantile position
    if likeInput < (np.quantile(Like,.40)):
        if likeInput < (np.quantile(Like,.20)):
            return "Worst"
        elif likeInput >= (np.quantile(Like,.20)):
            return "Bad"
    elif likeInput >= (np.quantile(Like,.40)):
        if likeInput < (np.quantile(Like,.60)) :
            return "Good"
        else:
            return "Best"

def coutLike(input):
    like = []
    for y in input['GraphImages']:
        like.append(y['edge_media_preview_like']['count'])
    return like
        
```
